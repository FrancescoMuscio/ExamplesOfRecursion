# Somma dei primi numeri dispari
Scrivere una funzione ricorsiva che calcoli la somma dei primi numeri dispari di n.

```python
def dispari(n):
    if(n==0): return 0
    if(n%2 == 0): return dispari(n-1)
    elif((n%2) != 0): return n + dispari(n-1)

print(dispari(3))
```

Scrivere una funzione ricorsiva che conti quanti elementi di un array A di lunghezza n sono minori di zero.

```python
def neg(A, n):
    if(n < 0): return 0
    if(A[n] > 0): return neg(A, n - 1)
    return 1 + neg(A, n - 1)

A = [-1, 2, -3, 4, -5, -6, 9, -15]
print(neg(A, len(A) - 1))
```

# Prodotto dei valori pari 
Scrivere una funzione ricorsiva che calcoli il prodotto di tutti gli elementi pari di un array A di lunghezza n. Se l’array non contiene elementi pari, la funzione deve restituire 1.

```python
def prod(A, n):
    if(n < 0): return 1
    if(A[n] % 2 != 0): return prod(A, n - 1)
    else: return A[n] * prod(A, n - 1)

A = [1, 3, 7, 5, 9, -2, 6, -10]
print(prod(A, len(A) - 1))
```

# Conteggio dei valori superiori alla media
Scrivere una funzione ricorsiva che conti quanti elementi di un array A di lunghezza n hanno valore superiore alla media aritmetica dell’array stesso.

```python
def med(A, n, media):
    if(n < 0): return 0
    if(A[n] < media): return med(A, n - 1, media)
    else: return 1 + med(A, n - 1, media)

A = [1, 3, 7, 5, 8]
media = sum(A)/len(A)
print(med(A, len(A) - 1, media))
```

# Differenza tra somma dei pari e somma dei dispari
Scrivere una funzione ricorsiva che calcoli la differenza tra la somma degli elementi pari e la somma degli elementi dispari di un array A di lunghezza n.

```python
def dif(A, n):
    if(n < 0): return 0
    if(A[n] % 2 != 0): return dif(A, n - 1) - A[n]
    else: return dif(A, n - 1) + A[n]

A = [1, 7, 8, 10, 2, 13]
print(abs(dif(A, len(A) - 1))) #il valore assoluto serve solo per non far stampare il negativo
```

```python
def simmetrico(A, i=0):
    if i >= len(A) // 2:
        return True
    if A[i] != A[len(A) - 1 - i]:
        return False
    return simmetrico(A, i + 1)

def simmetrico2(A, l, r):
    if(l >= r): return True
    if(A[l] != A[r]): return False
    else: return simmetrico2(A, l + 1, r - 1)


A = [13, 7, 8, 8, 7, 13]
print(simmetrico(A))
print(simmetrico2(A,0, len(A) - 1))
```

# Conteggio dei picchi locali
Scrivere una funzione ricorsiva che conti quanti elementi di un array A sono picchi locali, ossia valori tali che A[i] > A[i − 1] e A[i] > A[i + 1].

```python
def picchi(A, l, r, n = 1):
    if(n == len(A) - 1): return 0
    if(A[n] > A[l] and A[n] > A[r]): return 1 + picchi(A, l + 1, r + 1, n + 1)
    else: return picchi(A, l + 1, r + 1, n + 1)

A = [0, 2, 1, 7, 3, 8, 7, 13, 11]
print(picchi(A, 0, 2))
```

```python
def mediaR(A, n):
    if(n < 0): return 0
    else: return A[n]/(len(A)) + mediaR(A, n - 1)

A = [4, 3, 4, 5, 9]
print(mediaR(A, len(A) - 1))
```

# Verifica della presenza di una sottosequenza crescente 
Scrivere una funzione ricorsiva che verifichi se all’interno di un array A esiste una sequenza di almeno tre elementi consecutivi in ordine strettamente crescente. La funzione deve restituire true o false.

```python
def sequenza(A, l, n, r):
    if(len(A) < 3): return False
    if(n == len(A) - 1): return False
    if(A[l] == A[n] - 1 and A[r] == A[n] + 1): return True
    return sequenza(A, l + 1, n + 1, r + 1)

A = [1,2,4,5,6,10,11]
print(sequenza(A, 0, 1, 2))
```

# Conteggio dei valori univoci 
Scrivere una funzione ricorsiva che conti quanti elementi di un array A non si ripetono. Un elemento è detto univoco se compare una sola volta nell’array.

```python
def conta_univoci(A, n, i=0, j=0, count=0):
    if i == n:
        return 0

    if j == n:
        if count == 1:
            return 1 + conta_univoci(A, n, i + 1, 0, 0)
        else:
            return conta_univoci(A, n, i + 1, 0, 0)

    if A[j] == A[i]:
        return conta_univoci(A, n, i, j + 1, count + 1)
    else:
        return conta_univoci(A, n, i, j + 1, count)


A = [2,3,2,3,6,9,9]
print(conta_univoci(A, len(A)))
```

# BST

```python
class RBNode:
      # Constructor to initialize node of RB Tree
    def __init__(self, value, color='red'):
        self.value = value
        self.color = color
        self.left = None
        self.right = None
        self.parent = None

    # function to get the grandparent of node
    def grandparent(self):
        if self.parent is None:
            return None
        return self.parent.parent

    # function to get the sibling of node
    def sibling(self):
        if self.parent is None:
            return None
        if self == self.parent.left:
            return self.parent.right
        return self.parent.left

    # function to get the uncle of node
    def uncle(self):
        if self.parent is None:
            return None
        return self.parent.sibling()
```

```python
class RBTree:
    def __init__(self):
        self.root = None

    def left_rotate(self, x): 
        y = x.right
        x.right = y.left

        if y.left is not None:
            y.left.parent = x

        y.parent = x.parent

        if x.parent is None:
            self.root = y
        elif x == x.parent.left:
            x.parent.left = y
        else:
            x.parent.right = y

        y.left = x
        x.parent = y

    def right_rotate(self, x):
        y = x.left
        x.left = y.right

        if y.right is not None:
            y.right.parent = x

        y.parent = x.parent

        if x.parent is None:
            self.root = y
        elif x == x.parent.right:
            x.parent.right = y
        else:
            x.parent.left = y

        y.right = x
        x.parent = y

    def insert(self, value):
        new_node = RBNode(value)
        y = None
        x = self.root

        while x is not None:
            y = x
            if new_node.value < x.value:
                x = x.left
            else:
                x = x.right

        new_node.parent = y

        if y is None:
            self.root = new_node
        elif new_node.value < y.value:
            y.left = new_node
        else:
            y.right = new_node

        new_node.color = 'red'
        self.fix_insert(new_node)

    def fix_insert(self, node):
        while node.parent is not None and node.parent.color == 'red':
            if node.parent == node.grandparent().left:
                uncle = node.uncle()

                # Caso 1: zio rosso
                if uncle is not None and uncle.color == 'red':
                    node.parent.color = 'black'
                    uncle.color = 'black'
                    node.grandparent().color = 'red'
                    node = node.grandparent()
                else:
                    # Caso 2: triangolo
                    if node == node.parent.right:
                        node = node.parent
                        self.left_rotate(node)

                    # Caso 3: linea
                    node.parent.color = 'black'
                    node.grandparent().color = 'red'
                    self.right_rotate(node.grandparent())
            else:
                # simmetrico
                uncle = node.uncle()

                if uncle is not None and uncle.color == 'red':
                    node.parent.color = 'black'
                    uncle.color = 'black'
                    node.grandparent().color = 'red'
                    node = node.grandparent()
                else:
                    if node == node.parent.left:
                        node = node.parent
                        self.right_rotate(node)

                    node.parent.color = 'black'
                    node.grandparent().color = 'red'
                    self.left_rotate(node.grandparent())

        self.root.color = 'black'

    def inorder(self, node):
        if node is not None:
            self.inorder(node.left)
            print(f"{node.value} ({node.color})", end="  ")
            self.inorder(node.right)

    def search(self, value):
        current = self.root
        while current is not None:
            if value == current.value:
                return current
            elif value < current.value:
                current = current.left
            else:
                current = current.right
        return None
    
    def print_tree(self, node, indent="", last=True):
        if node is not None:
            print(indent, end="")
            if last:
                print("R----", end="")
                indent += "     "
            else:
                print("L----", end="")
                indent += "|    "

            print(f"{node.value} ({node.color})")
            self.print_tree(node.left, indent, False)
            self.print_tree(node.right, indent, True)

```

```python
tree = RBTree()
values = [10, 20, 30, 15, 25, 5, 3, 40, 42]

for v in values:
    tree.insert(v)

tree.print_tree(tree.root)
   
```

# Verifica di BST valido 
Scrivere una funzione ricorsiva che verifichi se un albero binario generico rispetta le proprietà di un BST. La funzione deve restituire true se per ogni nodo v tutti i valori nel sottoalbero sinistro sono minori di v, e tutti quelli nel sottoalbero destro sono maggiori.

```python
def check_bst(node, min_val=float('-inf'), max_val=float('inf')):
    if node is None:
        return True

    if not (min_val < node.value < max_val):
        return False

    return (check_bst(node.left, min_val, node.value) and
            check_bst(node.right, node.value, max_val))

check_bst(tree.root)
```

```python
def sum_path(node, k):
    if node is None: return False
    if(node.left is None and node.right is None): return node.value == k #serve per far ritornare il True
    remaining = k - node.value
    return sum_path(node.left, remaining) or sum_path(node.right, remaining)

sum_path(tree.root, 38)
```

# Verifica di bilanciamento 
Scrivere una funzione ricorsiva che verifichi se un BST è bilanciato, ossia se per ogni nodo la differenza tra le altezze dei due sottoalberi non supera 1. La funzione deve restituire true/false.

```python
def is_balanced(node):
    if node is None:
        return True, 0  # bilanciato, altezza 0

    left_bal, left_h = is_balanced(node.left)
    if not left_bal:
        return False, 0

    right_bal, right_h = is_balanced(node.right)
    if not right_bal:
        return False, 0

    if abs(left_h - right_h) > 1:
        return False, 0

    return True, 1 + max(left_h, right_h)

is_balanced(tree_1.root)
```

# Verifica della presenza di una sottostringa
Scrivere una funzione ricorsiva che determini se una stringa T è contenuta all’interno di una stringa S. La funzione deve restituire true o false.

```python
def contains(S, T):
    # T vuota è sempre contenuta
    if T == "":
        return True

    # S troppo corta
    if S == "":
        return False

    # confronto prefisso
    if S.startswith(T):
        return True

    # prova togliendo il primo carattere di S
    return contains(S[1:], T)

contains('pallavolo', 'volo')
```

# Conteggio delle parole 
Scrivere una funzione ricorsiva che calcoli quante parole contiene una stringa S, assumendo che le parole siano separate da uno spazio singolo.

```python
def words_count(S, i):
    if(i == len(S)): return 0
    if(S == ""): return 0
    if(i == 0 and S[i] != ' '): return 1 + words_count(S, i + 1)
    if(S[i] == ' '): return 1 + words_count(S, i + 1)
    return words_count(S, i + 1)

def words_count2(S, i=0):
    if i == len(S):
        return 0

    # inizio parola
    if i == 0 and S[i] != ' ':
        return 1 + words_count2(S, i + 1)

    if S[i] != ' ' and S[i-1] == ' ':
        return 1 + words_count2(S, i + 1)

    return words_count2(S, i + 1)


S = 'Ciao parola2 345 ola'
print(words_count(S, 0))
print(words_count2(S))
```
